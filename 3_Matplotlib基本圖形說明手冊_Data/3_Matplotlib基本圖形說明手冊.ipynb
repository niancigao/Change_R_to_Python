```
R 中的 abline 函數可用於向繪圖添加直線

不幸的是，這個函數在 Matplotlib 中不存在，
但我們可以定義以下函數來複製 Python 中的 abline 函數：
```

+ [如何在 Matplotlib 中使用 abline 函數](https://www.statology.org/matplotlib-abline/)


````python
def abline(slope, intercept):
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--')
````

### 直方圖 : plt.hist()
+ 直方圖（Histogram）是一組連續數字（Numerical）資料的次數分配圖
2. 載入資料
+ scikit-learn的波士頓住房數據集

````python
boston = pd.read_csv(r"C:\Users\user\Desktop\TW-04-301000000A-060043\Boston.csv")
boston.head(5)    # 顯示前 5 筆資料
````

3. **畫圖**
+ [matplotlib.pyplot.hist](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)
+ [命名顏色列表](https://matplotlib.org/stable/gallery/color/named_colors.html)

````python
plt.hist(boston["age"],                                       # 輸入值，這需要單個數組
         bins = 10,                                           # 分箱數，默認值10，也可放 'auto','fd','doane','scott','stone','rice','sturges','sqrt'
         range = (boston["age"].min(), boston["age"].max()),  # 箱的下限和上限範圍
         density = True,                                      # True，繪製並返回概率密度
         weights = None,                                      # 權重數組，與 boston["age"] 具有相同的形狀。如果 density = True，則對權重進行歸一化，以便密度在範圍內的積分保持為 1
         cumulative = False,                                  # 如果 True 則計算累積直方圖，最後一個 bin 給出了數據點的總數，如果為 -1，則累積的方向相反
         bottom = 1,                                          # 每個箱子底部的位置(看 y 軸)，bins are drawn from bottom to bottom + hist(x, bins)
         histtype = 'bar',                                    # 'bar' 條形圖並排排列，'barstacked' 多個數據堆疊在一起，'step' 生成未填充的線圖，'stepfilled' 生成一個默認填充的線圖
         align = 'mid',                                       # 水平對齊方式，'left'：以左 bin 邊緣為中心，'mid'：在 bin 邊緣之間居中，'right'：以右 bin 邊緣為中心
         orientation = 'vertical',                            # {'vertical', 'horizontal'} 垂直或水平的直方圖
         rwidth = 0.95,                                       # 條形的相對寬度作為 bin 寬度的一部分，0-1
         log = False,                                         # 如果 True，直方圖軸將設置為對數刻度
         color = '#7cb5ec',                                   # 可參考上面 命名顏色列表 連結
         label = "Age",                                       # str. 只有一個標籤，legend will work as expected
         stacked = False,                                     # 如果為 True，則多個數據相互堆疊(重圖) 如果為 False，則多個數據並排排列
         data=None)

plt.legend(loc = "best")                                      # 如果有 label 則要放
plt.show()
````

### Pair Plot
```
由於 Matplotlib 比較原始，沒有提供現成的 pair-plot 函數，我們可以像 Seaborn 那樣自己做，
如果 Seaborn 等模塊中已經提供此類自製函數，您通常不必創建它們，
但是以自定義方式實施您的可視化方法讓您有機會了解您繪製的內容，有時可能與現有方法有很大不同，
在這裡使用 Matplotlib 創建我們的 pair-plot
```
+ [使用 Matplotlib 自定義配對圖](https://www.endpointdev.com/blog/2022/04/visualizing-data-with-pairplot-using-matplotlib/)

````python
### 最初我們需要使用 matplotlib 的 subplots 函數創建一個網格繪圖區域
# fig, axis = plt.subplots(nrows=3, ncols=3)

### 為我們創建的繪圖網格區域準備一個繪圖函數
### 如果我們的數據集中有 3 個特徵，如本例所示，我們可以像這樣遍歷每個特徵的特徵

"""
for i in range(0, 3):
    for j in range(0, 3):
        plotPair()
"""

### single pair plotting to another function
### 在網格中繪製單個對需要的數據
### 統一將標籤繪製在網格區域的最左側和最底部
def plot_single_pair(ax, feature_ind1, feature_ind2, _X, _y, _features, colormap):
    """Plots single pair of features.

    Parameters
    ----------
    ax : Axes
        matplotlib axis to be plotted
    feature_ind1 : int
        index of first feature to be plotted
    feature_ind2 : int
        index of second feature to be plotted
    _X : numpy.ndarray
        Feature dataset of of shape m x n
    _y : numpy.ndarray
        Target list of shape 1 x n
    _features : list of str
        List of n feature titles
    colormap : dict
        Color map of classes existing in target

    Returns
    -------
    None
    """

    # 如果特徵相同（成對圖的對角線），則繪製分佈直方圖
    if feature_ind1 == feature_ind2:
        tdf = pd.DataFrame(_X[:, [feature_ind1]], columns = [_features[feature_ind1]])
        tdf['target'] = _y
        for c in colormap.keys():
            tdf_filtered = tdf.loc[tdf['target']==c]
            ax[feature_ind1, feature_ind2].hist(tdf_filtered[_features[feature_ind1]], color = colormap[c], bins = 30)
    else:
        # 否則繪製成對散點圖
        tdf = pd.DataFrame(_X[:, [feature_ind1, feature_ind2]], columns = [_features[feature_ind1], _features[feature_ind2]])
        tdf['target'] = _y
        for c in colormap.keys():
            tdf_filtered = tdf.loc[tdf['target']==c]
            ax[feature_ind1, feature_ind2].scatter(x = tdf_filtered[_features[feature_ind2]], y = tdf_filtered[_features[feature_ind1]], color=colormap[c])

    # 僅在配對圖的左側/和底部打印特徵標籤
    # 這裡避免在內圖打印標籤
    if feature_ind1 == len(_features) - 1:
        ax[feature_ind1, feature_ind2].set(xlabel=_features[feature_ind2], ylabel='')
    if feature_ind2 == 0:
        if feature_ind1 == len(_features) - 1:
            ax[feature_ind1, feature_ind2].set(xlabel=_features[feature_ind2], ylabel=_features[feature_ind1])
        else:
            ax[feature_ind1, feature_ind2].set(xlabel='', ylabel=_features[feature_ind1])

````

````python
###　成對圖的最終網格圖函數：

def myplotGrid(X, y, features, colormap={0: "red", 1: "green", 2: "blue"}):
    """Plots a pair grid of the given features.

    Parameters
    ----------
    X : numpy.ndarray
        Dataset of shape m x n
    y : numpy.ndarray
        Target list of shape 1 x n
    features : list of str
        List of n feature titles

    Returns
    -------
    None
    """

    feature_count = len(features)
    # 創建一個大小為 [feature count x feature count] 的 matplot 子圖區域
    fig, axis = plt.subplots(nrows=feature_count, ncols=feature_count)
    # 設置圖形大小有助於根據特徵數優化圖形大小
    fig.set_size_inches(feature_count * 4, feature_count * 4)

    # 遍歷特徵以繪製成對圖
    for i in range(0, feature_count):
        for j in range(0, feature_count):
            plot_single_pair(axis, i, j, X, y, features, colormap)

    plt.show()
````

### 3D圖 : 多邊形

1. **載入套件**

````python
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
````

3. **畫圖**
+ 立方體

````python
width = 30
depth = 40
height = 50

def box(width, depth, height):
    faces = Poly3DCollection([
        [[0, 0, 0], [width, 0, 0], [width, depth, 0], [0, depth, 0]],
        [[0, 0, height], [width, 0, height], [width, depth, height], [0, depth, height]],
        [[0, 0, 0], [width, 0, 0], [width, 0, height], [0, 0, height]],
        [[0, depth, 0], [width, depth, 0], [width, depth, height], [0, depth, height]],
        [[0, 0, 0], [0, depth, 0], [0, depth, height], [0, 0, height]],
        [[width, 0, 0], [width, depth, 0], [width, depth, height], [width, 0, height]]
    ])
    faces.set_edgecolor('black')

    ax = plt.axes(projection='3d')
    ax.add_collection3d(faces)

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')

    lim = max(width, depth, height)
    ax.set_xlim([0, lim])
    ax.set_ylim([0, lim])
    ax.set_zlim([0, lim])

    plt.show()

box(width, depth, height)
````

3. **畫圖**
+ 正四面體

````python
def tetrahedron(width):
    n = width / (2 ** 0.5) * 0.5;

    xs = np.array([n, -n,  n, -n]) 
    ys = np.array([n,  n, -n, -n]) 
    zs = np.array([n, -n, -n, n]) 

    coord = np.dstack((xs, ys, zs))[0]

    faces = Poly3DCollection([
        coord[[0, 1, 2]], 
        coord[[1, 2, 3]], 
        coord[[2, 3, 0]], 
        coord[[3, 0, 1]]
    ])
    faces.set_edgecolor('black')

    ax = plt.axes(projection='3d')
    ax.add_collection3d(faces)

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')

    ax.set_xlim([-n, n])
    ax.set_ylim([-n, n])
    ax.set_zlim([-n, n])

    plt.show()

width = 30
tetrahedron(width)

plt.savefig("名字.jpg",                          # 儲存圖檔，文件格式，例如 “png”、“pdf”、“svg”…
            bbox_inches = 'tight',               # 去除座標軸占用的空間
            pad_inches = 0.0,                    # 去除所有白邊
            facecolor='auto',                    # 顏色，如果為 “自動”，則使用當前圖形面的顏色
            edgecolor='auto')                    # 圖的邊緣顏色，如果為 “自動”，則使用當前圖形邊緣顏色
````
